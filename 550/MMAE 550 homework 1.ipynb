```python
import numpy as np
import matplotlib.pyplot as plt

die = [(1/6),(1/6),(1/6),(1/6),(1/6),(1/6)]
dice = die

for n in range(2,5):
  dice = np.convolve(dice,die)
  sums = []
  for i in range(n, 6*n+1):
    sums.append(i)
  plt.figure()
  plt.bar(sums,dice)
  plt.xlabel("Sum of dice showing")
  plt.ylabel("Probability")
  plt.show()

#print(sums)
```
