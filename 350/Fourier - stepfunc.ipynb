```python
import numpy as np
import matplotlib.pyplot as plt

resolution = 0.0001
x = np.arange(-np.pi,np.pi,resolution)

#y==square; t==x
square =  np.zeros_like(x)
squarefft =  np.zeros_like(x)
#Ak=(2/(k*pi))*(1-(-1)^k)
#SUM(Ak*sin(kx)); k=0 to n

square[:int(x.size/2)] = -1.
square[int(x.size/2):]= 1.

n=10000
for k in range(n):
    a=2/((k+1)*np.pi)*(1-(-1)**(k+1))
    squarefft=squarefft+a*np.sin((k+1)*x)

# plt.plot(x,square)    
plt.plot(x,square,x,squarefft)
```

```python
#Fourier sine and cosine series
resolution = 0.0001
x = np.arange(-np.pi,np.pi,resolution)
square =  np.zeros_like(x)
squarefft =  np.zeros_like(x)
square[:int(x.size/2)] = 1.
square[int(x.size/2):]= 0

# plt.plot(x,square)

a0=np.trapz(square,x)/np.pi/2
squarefft=squarefft+a0
n=5
for i in range(n):
    i=i+1
    a=np.trapz(square*np.sin(i*x),x)/np.pi
    b=np.trapz(square*np.cos(i*x),x)/np.pi
    squarefft=squarefft+a*np.sin(i*x)+b*np.cos(i*x)
    
plt.plot(x,square,x,squarefft)
```
