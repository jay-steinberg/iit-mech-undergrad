```python
import numpy as np  # also can pull from scipy
import matplotlib.pyplot as plt
from scipy.io import savemat

fs=500 # sampling frequency, nyquist freq is fs/2
dt=1/fs # time discretization
N=500 # number of data points
T=N*dt # max time in the data, essentially tau
freq = np.linspace(0, fs, N)  # this is [f0 f1 f2 etc]
t = np.linspace(0,T,N) # time vector [t1 t2 t3 etc]

print(dt, fs)

f1, f2, f3, f4, f5 = 50, 100, 150, 200, 25
s = np.sin(f1 * 2 * np.pi * t) + 0.5 * np.sin(f2 * 2 * np.pi * t) + np.sin(f3 * 2 * np.pi * t) + 0.5 * np.sin(f4 * 2 * np.pi * t) + 0.5 * np.sin(f5 * 2 * np.pi * t)

plt.figure()
plt.ylabel('Amplitude')
plt.xlabel('Time [s]')
plt.plot(t,s)
plt.show()
```

```python
f = np.fft.fft(s)

plt.figure()
plt.ylabel('Amplitude')
plt.xlabel('Frequency [Hz]')
plt.plot(freq[:N//2], np.abs(f)[:N//2]*1/N)  # // double divided gives modulus division, or integer return only instead of float point
plt.show()
```
