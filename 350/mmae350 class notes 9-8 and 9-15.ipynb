```python
ax = plt.axes(projection="3d")

def z_function(x, y):
  a=2
  b=3
  return (np.sqrt(a*x**2 + b*y**2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

#creating a grid from vectors
X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)
print(np.shape(X))
print(X[0,0])

ax.plot_surface(X, Y, Z, cmap='jet')
ax.set_title('surface')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
