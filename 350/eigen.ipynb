```python
import numpy as np
from numpy import linalg as la

a = np.array([[6,3],[3,5]])
print(a, '\n')
eigenvalues, eigenvectors = la.eig(a)
print(eigenvalues, '\n')  # the returned eigenvalues are not sorted in any particular manner
print(eigenvectors, '\n') # output of eigenvectors is a row of column eigenvectors (x0 is first column)
print(la.norm(eigenvectors[:,0])) # length of eigenvectors are 1 (eigenvectors are always normalized)
print(la.norm(eigenvectors[:,1]))

e1 = eigenvectors[:,0]
e2 = eigenvectors[:,1]

print(np.dot(e1,e2)) # dot product is zero, the eigenvectors are perpendicular for any symmetric matrix

```

```python
a = np.random.rand(3,3) # random matrix
print (a)
a=a+a.T # sum with transpose to make symmetric
print(a)

v, e = la.eig(a)

print(v)
print('First Eigenvector:')
print(e[:,0])
print('Second Eigenvector:')
print(e[:,1])
print('Third Eigenvector:')
print(e[:,2])

print(np.dot(e[:,0], e[:,1]))
print(np.dot(e[:,1], e[:,2]))
```
