```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import scipy.optimize as op
```

```python
#ax^2 + bx + c = 0

a,b,c = 1.,4.,1.

x1=(-b+np.sqrt(b**2-4*a*c))/(2*a)
x2=(-b-np.sqrt(b**2-4*a*c))/(2*a)

print('The two solutions are:')
print(x1,x2)

x = np.arange(-5.,2.,.01)
y = a*x**2+b*x+c 

# plt.xlim([-.3,-.2])
# plt.ylim([-.2,.2])

plt.plot(x,y,x,x*0)
```

```python
#one equation

def oneEquation(z):  # defining f(z) assuming z is a vector containing one variable x located at z[0]

#if f(x,y) = 0 you can still pass all vars -> z[0] = x, z[1] = y

  a,b,c=1.,4.,1.
  x = z[0]
  F = np.empty((1))  # save soln in vector F of size 1
  F[0] = a*x**2+b*x+c 
  return F

zGuess = np.array([-4])
z = fsolve(oneEquation,zGuess)  #  for fsolve, first argument is the function to solve, second is guess
print(z) #fsolve solution near guess
print(oneEquation(z)) #checking fsolve solution back into original func

```

```python
#minimization and solution of nonlinear equations
#find the minima of f(x)=0.1x^3+3x&2+5x+4

x = np.arange(-10.,10.,.01)
y=0.1*x**3+3*x**2+5*x+4

plt.plot(x,y)

def f(z):
  x = z[0]
  F= np.empty((1))
  F[0] = 0.1*x**3+3*x**2+5*x+4
  return F

#minimize original eqn

zGuess = np.array([0])
z = op.minimize(f,zGuess)
print(z)

# or solve derivative = 0

plt.figure()
yp = 0.3*x**2+6*x+5
plt.plot(x,yp,x,0*x) 

def dfdx(z):
  x = z[0]
  F = np.empty((1))
  F[0] = 0.3*x**2+6*x+5
  return F

zGuess = np.array([0])
z = fsolve(dfdx, zGuess)
print(z)
```
