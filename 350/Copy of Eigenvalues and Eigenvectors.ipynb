```python
import numpy as np
from numpy import linalg as la
```

```python
a=np.array([[6,3],[3,5]])
print(a)
eigenvalues,eigenvectors=la.eig(a)
print(eigenvalues)
print(eigenvectors)
print(la.norm(eigenvectors[:,0]))
print(la.norm(eigenvectors[:,1]))
e1 = eigenvectors[:,0]
e2 = eigenvectors[:,1]
print(np.dot(e1,e2))
```

```python
a=np.random.rand(3,3)
# print(a)
a=a+a.T
print(a)

v,e=la.eig(a)
print(v)
print('Eigenvectors:')
print('First Eigenvector:')
print(e[:,0])

print('Second Eigenvector:')
print(e[:,1])

print('Third Eigenvector:')
print(e[:,2])

print(np.dot(e[:,0],e[:,1]))
print(np.dot(e[:,1],e[:,2]))
```

```python
q,r=la.qr(a)
a=r@q
print(a)
```
