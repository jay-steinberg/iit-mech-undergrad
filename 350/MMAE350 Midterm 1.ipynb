## Q2:

Solve the following system of linear equations for unknowns $x1-x4$:

eqns.png

```python
coeffs = np.array([[2, 1, -5, -7], 
                   [1, -2, 0, -8],
                   [2, -7, 2, -5],
                   [-1, 9, 1, 8]])

constants = np.array([10, -5, 0, 21])

X = np.linalg.inv(coeffs)@constants

print("Original system coefficients: \n", coeffs)
print("Original system constants ", constants)
print("Solutions for x: ", X)

newconsts = np.round(coeffs@X, decimals=5)

print("Solutions substituted into system yields constants: ", newconsts)
print("Does the substitution match? ", newconsts == constants)
```

## Q3:

Consider $y = cos^2(x)$. Write down the Taylor series approximation $y_a(x)$ of $y(x)$ near $x_0 = \pi /4$ including only the first two terms. Plot both $y(x), y_a(x)$ as a function of $x$ near $x_0$.

> The Taylor expansion is initially done by hand for $n$ terms:

> $T(x) = y(x_o) + (x-x_o) \frac{dy}{dx}|_{x_o}) + ... $

```python
x = np.linspace(0, np.pi/2, 50) 
yfunc = (np.cos(x)**2)

x0 = np.pi/4

y2 = ((np.cos(x0)**2) + (x-x0)*(-2*np.cos(x0)*np.sin(x0)))

plt.plot(x,yfunc, label='y')
plt.plot(x, y2,label='y2')
plt.legend()
plt.show()
```

## Q4:

Consider function $y(x) = x^3 + 2x^2 - 12x + 20$

a) plot $y(x)$ as a function of $x$ for $x$ between $-5,4$.

b) Find the minimum value of $y(x)$ in this region.

C) At what value of $x$ is the minimum achieved?


```python
x = np.linspace(-5,4, 36)

def yfunc(z):
  F = np.empty((1))
  F = (z**3 + 2*z**2 - 12.*z + 20)
  return F 

plt.plot(x, yfunc(x))
plt.show()
print("By visual inspection, the minimum value of Y on this whole region is at the lower bound x=-5, y = ", yfunc(-5), '\n')

minX = op.minimize(yfunc, 1).x
print("A local minimum can also be found at x = ", minX, ", y = ", yfunc(minX))
```
